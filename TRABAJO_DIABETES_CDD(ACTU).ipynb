# CRISP-DM 1: Comprension del negocio


Introducción

La diabetes es una enfermedad crónica que se presenta cuando los niveles de glucosa en la sangre (comúnmente conocida como azúcar) son demasiado altos. La glucosa es vital, ya que es la principal fuente de energía de nuestro cuerpo y proviene tanto de los alimentos que consumimos como de la producción interna del propio organismo.

Para que esta azúcar pueda entrar a las células y convertirse en energía, el cuerpo necesita una hormona llamada insulina, producida por el páncreas. Cuando se tiene diabetes, este proceso se rompe: el cuerpo no produce insulina, no produce la cantidad suficiente o no la utiliza de manera correcta. Como resultado, la glucosa se acumula en la sangre en lugar de alimentar a las células.

¿Qué es la diabetes y cuales hay?

Aunque existen variantes como la diabetes gestacional (que ocurre durante el embarazo), los dos tipos principales son:

- Diabetes Tipo 1: El sistema inmunitario ataca por error y destruye las células del páncreas que fabrican la insulina, provocando que el cuerpo produzca muy poca o ninguna. Suele diagnosticarse en niños y adultos jóvenes, y quienes la padecen necesitan administrarse insulina todos los días para vivir.

- Diabetes Tipo 2: Es el tipo más común. Aquí, las células no usan la insulina correctamente (resistencia) o el páncreas no produce suficiente. Aunque puede aparecer a cualquier edad, el riesgo aumenta considerablemente debido a factores como el sobrepeso, la obesidad y el historial familiar.

## Diccionario de Salud Metabólica

- **Glucosa estable**

Qué es: El nivel de azúcar en la sangre tras un ayuno.

Cómo se mide: A través de una muestra de sangre (mg/dL).

Cómo afecta la diabetes: Es el indicador directo; si es alta de forma constante, confirma la incapacidad del cuerpo para procesar la insulina.

Valores: Bajo: <70 mg/dL | Normal: 70–99 mg/dL | Alto: ≥126 mg/dL (Criterio de diabetes).



- **Hemoglobina glicosilada**

Qué es: El promedio de azúcar en los últimos 3 meses.

Cómo se mide: Análisis de sangre que mide el porcentaje de glucosa pegada a los glóbulos rojos.

Cómo afecta la diabetes: Determina si el tratamiento actual está funcionando o si el azúcar ha estado fuera de control a largo plazo.

Valores: Bajo: <5.7% | Normal: 5.7%–6.4% | Alto: ≥6.5% (Criterio de diabetes).



- **Colesterol**

Qué es: Grasa total presente en la sangre.

Cómo se mide: Análisis de sangre (mg/dL).

Cómo afecta la diabetes: La diabetes acelera el endurecimiento de las arterias; tener el colesterol alto aumenta drásticamente el riesgo de infartos.

Valores: Bajo: <200 mg/dL | Normal: 200–239 mg/dL | Alto: ≥240 mg/dL.


- **HDL**

Qué es: Proteína conocida como colesterol "bueno" que transporta el colesterol al hígado para eliminarlo.

Cómo se mide: Análisis de sangre (mg/dL).

Cómo afecta la diabetes: Niveles bajos son comunes en pacientes diabéticos y aumentan el riesgo cardiovascular.

Valores: Bajo: <40 mg/dL | Normal: 40–60 mg/dL | Alto: >60 mg/dL.


- **Ratio colesterol hdl**

Qué es: La relación matemática entre el colesterol total y el HDL.

Cómo se mide: División (Colesterol total ÷ HDL).

Cómo afecta la diabetes: Indica qué tan "sucia" está la sangre; un ratio alto significa mayor riesgo de obstrucción en las arterias.

Valores: Bajo: <3.5 | Normal: 3.5–5.0 | Alto: >5.0.

Presión sistolica 1/2

Qué es: La fuerza de la sangre contra las arterias cuando el corazón late.

Cómo se mide: Con un tensiómetro (mmHg).

Cómo afecta la diabetes: La diabetes daña los vasos sanguíneos, lo que eleva la presión; mantenerla controlada es vital para evitar complicaciones.

Valores: Bajo: <90 mmHg | Normal: 90–120 mmHg | Alto: ≥140 mmHg.


- **Presión diastolica 1/2**

Qué es: La fuerza de la sangre contra las arterias cuando el corazón descansa entre latidos.

Cómo se mide: Con un tensiómetro (mmHg).

Cómo afecta la diabetes: Al igual que la sistólica, su elevación aumenta el daño crónico a los órganos internos.

Valores: Bajo: <60 mmHg | Normal: 60–80 mmHg | Alto: ≥90 mmHg.

Cintura

Qué es: Medida de la circunferencia abdominal.

Cómo se mide: Cinta métrica a la altura del ombligo (cm).

Cómo afecta la diabetes: La grasa abdominal libera sustancias que generan resistencia a la insulina, dificultando el control de la glucosa.

Valores: Bajo: No aplica | Normal: <88 cm (mujeres) / <102 cm (hombres) | Alto: ≥88 cm (mujeres) / ≥102 cm (hombres).

¿Cuáles son los síntomas y cómo identificarlos?
Cuando el azúcar comienza a acumularse en la sangre, el cuerpo envía señales claras. Los principales síntomas de la diabetes incluyen:

- Aumento notable de la sed y necesidad frecuente de orinar (especialmente por la noche).

- Pérdida de peso inexplicable y mucha hambre o cansancio extremo.

- Problemas físicos como visión borrosa, piel muy seca, hormigueo o entumecimiento en manos y pies.

- Dificultades de recuperación, como llagas que tardan mucho en cicatrizar o tener más infecciones de lo habitual.

¿Cómo prevenir la diabetes?

a posibilidad de prevenir la diabetes depende enteramente del tipo del que estemos hablando, ya que sus causas son completamente distintas:

1. Diabetes Tipo 1: El enfoque en la detección temprana y la ciencia
Dado que la diabetes tipo 1 es una enfermedad autoinmunitaria (donde el cuerpo se ataca a sí mismo por error) influenciada por la genética y posibles desencadenantes ambientales como los virus, actualmente no existe una forma de prevenirla mediante la dieta o el ejercicio. Los hábitos de vida no la causan ni la evitan.

Sin embargo, el panorama médico está cambiando:

- Pruebas de anticuerpos: Si una persona tiene antecedentes familiares directos, los médicos pueden realizar análisis de sangre específicos para detectar si el sistema inmunitario ya ha comenzado a atacar las células del páncreas, mucho antes de que aparezcan los síntomas.

- Avances médicos: La investigación científica actual ha desarrollado tratamientos biológicos innovadores (como inmunoterapias) capaces de retrasar la aparición clínica de la diabetes tipo 1 por varios años en personas de alto riesgo, permitiendo ganar un tiempo valioso y preparar un mejor manejo de la salud.

2. Diabetes Tipo 2: El poder del estilo de vida
A diferencia de la tipo 1, la diabetes tipo 2 sí se puede retrasar o prevenir en la mayoría de los casos modificando ciertos hábitos. Si tienes factores de riesgo como sobrepeso u obesidad, las medidas clave son:

- Control del peso: Perder tan solo entre el 5% y el 7% de tu peso actual (por ejemplo, bajar de 10 a 14 libras si pesas 200) y mantenerlo a largo plazo disminuye drásticamente el riesgo.

- Alimentación saludable: Reducir calorías, grasas y azúcares mediante porciones más pequeñas. Se debe priorizar el consumo de granos integrales, frutas y verduras, limitando las carnes rojas y evitando las procesadas.

- Actividad física regular: Hacer al menos 30 minutos de ejercicio la mayoría de los días de la semana (como caminar a paso rápido, andar en bicicleta o nadar) ayuda a que tu cuerpo use la insulina de manera más eficiente y baje los niveles de azúcar en sangre.

- Evitar el tabaco: No fumar es clave, ya que el cigarrillo contribuye directamente a generar resistencia a la insulina.

###1. Presión Arterial 1: Reposo.

Qué es: Es la presión de tu sangre cuando estás tranquilo y sin moverte.

Por qué importa: Establece tu nivel base de salud. Se considera normal estar por debajo de 120/80.

###2. Presión Arterial 2: Actividad.

Qué es: Es la presión de tu sangre mientras realizas ejercicio físico.

Por qué importa: Ayuda a conocer tus límites de seguridad durante el esfuerzo físico, diferenciando entre lo que es normal en ejercicio y lo que podría considerarse una presión crítica.

###3. Control del Pulso.

Qué es: Mide qué tan rápido late tu corazón.

Por qué importa: Sirve como un indicador de riesgo arrítmico. Lo óptimo en reposo es entre 60 y 80 latidos por minuto (lpm); si supera los 100 lpm, se considera alto.

###4. Variabilidad de la Presión.

Qué es: Es la diferencia entre tu presión en actividad y tu presión en reposo.

Por qué importa: Evalúa tu estabilidad ante diferentes situaciones. Una diferencia menor a 10 es estable, mientras que una diferencia mayor a 20 en la presión sistólica (o 10 en la diastólica) indica mayor variabilidad.

###5. Tiempo PPN (Post-Prandial Numérico)

Qué es: Mide cuánto tiempo tarda tu cuerpo en recuperar niveles normales de glucosa después de comer.

Por qué importa: Funciona como un indicador de eficiencia insulínica. Un tiempo de recuperación rápido (1 hora) es óptimo, mientras que un tiempo prolongado (3 horas o más) puede indicar una falla en la gestión de la glucosa.

## Analisis clinico para responder las preguntas de negocio

En las siguientes celdas se crea una clasificacion academica de `diabetes_probable` usando la hemoglobina glicosilada y se calculan promedios de glucosa, peso, cintura, presion, edad y hemoglobina glicosilada.


```python
# Primero, eliminamos las filas donde falta el dato de hemoglobina,
# ya que si no tenemos el valor, no podemos clasificar al paciente.
# Usamos .copy() para trabajar sobre una copia y no alterar la tabla original.
datos_preguntas = df.dropna(subset=["hemoglobina_glicosilada"]).copy()

# Creamos la columna 'diabetes_probable' y le ponemos 0 a todos por defecto.
# Esto es como decir: "asumimos que no tienen diabetes hasta que se demuestre lo contrario".
datos_preguntas["diabetes_probable"] = 0

# Aquí aplicamos la regla médica: si el valor de hemoglobina es 6.5 o más,
# cambiamos el 0 por un 1 en esa fila específica.
datos_preguntas.loc[
    datos_preguntas["hemoglobina_glicosilada"] >= 6.5,
    "diabetes_probable"
] = 1

# Finalmente, contamos cuántas personas quedaron en cada grupo (0 y 1).
# Esto nos da un resumen rápido de cuántas personas tienen niveles sospechosos
# de diabetes frente a las que no.
datos_preguntas["diabetes_probable"].value_counts()
```

```python
# Agrupamos los datos en dos bloques: los que tienen 'diabetes_probable' = 0
# y los que tienen = 1. Al agrupar, preparamos el terreno para hacer cálculos
# matemáticos específicos para cada grupo.
resumen_clinico = datos_preguntas.groupby("diabetes_probable")[
    [
        # Seleccionamos solo las columnas de interés clínico.
        # Esto evita que el código intente promediar datos irrelevantes.
        "glucosa_estable",
        "peso",
        "cintura",
        "presion_sistolica_1",
        "edad",
        "hemoglobina_glicosilada"
    ]
# Calculamos el promedio (.mean()) para cada columna en cada grupo,
# y redondeamos a un decimal (.round(1)) para que la tabla sea fácil de leer.
].mean().round(1)

# Mostramos el resultado final.
resumen_clinico
```
